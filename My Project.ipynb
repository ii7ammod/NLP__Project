```python
def sentence_endings(text):
    if not isinstance(text, str):
        return []
    sentences = re.split(r'(?<=[\.!\؟\!؛])\s*', text)
    result = []
    for s in sentences:
        s = s.strip()
        if not s:
            continue
        last_char = s[-1]
        if last_char.isalnum():
            ending = "NO_PUNCTUATION"
        else:
            ending = last_char
        result.append(ending)
    return result
```

```python
df['q_sentence_endings'] = df['q_body'].apply(sentence_endings)
df['a_sentence_endings'] = df['a_body'].apply(sentence_endings)
```

```python
import torch
from transformers import (
    MT5ForConditionalGeneration,
    AutoTokenizer,
    Seq2SeqTrainingArguments,
    Seq2SeqTrainer,
    DataCollatorForSeq2Seq
)
from datasets import Dataset
from sklearn.model_selection import train_test_split
from huggingface_hub import snapshot_download
```

```python
MAX_INPUT_LENGTH = 512
MAX_TARGET_LENGTH = 256
BATCH_SIZE = 4
GRADIENT_ACCUMULATION_STEPS = 2
LEARNING_RATE = 5e-7
NUM_EPOCHS = 3
OUTPUT_DIR = "./mt5_arabic_medical_qa"
```

```python
# Convert to Hugging Face Dataset
train_dataset = Dataset.from_pandas(train_df[['q_body', 'a_body']])
val_dataset = Dataset.from_pandas(val_df[['q_body', 'a_body']])
test_dataset = Dataset.from_pandas(test_df[['q_body', 'a_body']])
```

```python
#Calling the model and preparing the tokenizer
print("Loading mT5 model and tokenizer...")


import os
os.environ['TRANSFORMERS_OFFLINE'] = '0'


try:
    # First try to load from cache
    tokenizer = AutoTokenizer.from_pretrained(
        MODEL_NAME, 
        use_fast=False,
        local_files_only=False
    )
    model = MT5ForConditionalGeneration.from_pretrained(
        MODEL_NAME,
        local_files_only=False
    )
    print("Loaded from cache successfully!")
except:
    # If not in cache, download model files
    print("Downloading model (this may take a few minutes)...")
    from huggingface_hub import snapshot_download
    
    
    cache_dir = snapshot_download(
        repo_id=MODEL_NAME,
        ignore_patterns=["*.msgpack", "*.h5", "*.ot", "*.safetensors"],
    )
    
    
    tokenizer = AutoTokenizer.from_pretrained(
        cache_dir,
        use_fast=False
    )
    model = MT5ForConditionalGeneration.from_pretrained(cache_dir)
    print("Download and loading completed!")
```

```python
# Preprocessing function
def preprocess_function(examples):
    """Tokenize questions and answers"""
    # Tokenize inputs (questions)
    model_inputs = tokenizer(
        examples['q_body'],
        max_length=MAX_INPUT_LENGTH,
        truncation=True,
        padding=False
    )
    
    # Tokenize targets (answers)
    labels = tokenizer(
        examples['a_body'],
        max_length=MAX_TARGET_LENGTH,
        truncation=True,
        padding=False
    )
    
    model_inputs['labels'] = labels['input_ids']
    return model_inputs
```

```python
# Apply preprocessing
print("Preprocessing datasets...")
train_dataset = train_dataset.map(
    preprocess_function,
    batched=True,
    remove_columns=['q_body', 'a_body']
)

val_dataset = val_dataset.map(
    preprocess_function,
    batched=True,
    remove_columns=['q_body', 'a_body']
)

test_dataset = test_dataset.map(
    preprocess_function,
    batched=True,
    remove_columns=['q_body', 'a_body']
)
```

```python
# Data collator for dynamic padding
data_collator = DataCollatorForSeq2Seq(
    tokenizer=tokenizer,
    model=model,
    padding="max_length",
    max_length=512
)
```

```python
# Training arguments
training_args = Seq2SeqTrainingArguments(
    output_dir=OUTPUT_DIR,
    eval_strategy="steps",
    eval_steps=1500,
    learning_rate=LEARNING_RATE,
    per_device_train_batch_size=BATCH_SIZE,
    per_device_eval_batch_size=BATCH_SIZE,
    gradient_accumulation_steps=GRADIENT_ACCUMULATION_STEPS,
    num_train_epochs=NUM_EPOCHS,
    weight_decay=0.01,
    save_total_limit=3,
    save_steps=3000,
    logging_steps=1000,
    predict_with_generate=True,
    fp16=False,
    push_to_hub=False,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    report_to="none",
    warmup_steps=100,  
    max_grad_norm=1.0,  
    dataloader_num_workers=0,  
)
```

```python
# Initialize trainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=data_collator,
    processing_class=tokenizer
)
```

```python
trainer.train(resume_from_checkpoint=CHECKPOINT_PATH)
```

```python
# Define inference function
def generate_answer(question, max_length=512):
    """Generate answer for a given question"""
    input_text = 'أجب عن السؤال: ' + question
    input_ids = tokenizer(
        input_text,
        return_tensors="pt",
        max_length=MAX_INPUT_LENGTH,
        truncation=True
    ).input_ids
    
    if torch.cuda.is_available():
        input_ids = input_ids.to('cuda')
        model.to('cuda')
    
    outputs = model.generate(
        input_ids,
        max_length=max_length,
        num_beams=4,
        early_stopping=True,
        no_repeat_ngram_size=2
    )
    
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return answer
```

```python
test_df = test_df.sample(frac=1).reset_index(drop=True)
```

```python
# Save test set predictions
print("\nGenerating predictions on test set...")
test_predictions = []
test_references = []

for i, example in enumerate(test_df.itertuples()):

    if i == 5:
        break
    
    question = example.q_body.replace('أجب عن السؤال: ', '')  
    reference = example.a_body
    prediction = generate_answer(question)
    
    test_predictions.append(prediction)
    test_references.append(reference)
    
    # Print first 5 examples
    if i < 5:
        print(f"\n--- Example {i+1} ---")
        print(f"Question: {question}")
        print(f"Reference: {reference}")
        print(f"Prediction: {prediction}")
```
